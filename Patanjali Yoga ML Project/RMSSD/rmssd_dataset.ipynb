```python
import pandas as pd
```

```python
demo = pd.read_csv('../Dataset - Demographic.csv')
demo.head()
```

```python
cols = demo.columns
nums = [1,9,10,11,13,14,15,16]

for i in nums:
    demo.drop(cols[i],axis=1,inplace=True)

demo = demo.rename(columns={'Subject_Id': 'ID'})
demo.fillna(demo.mean(numeric_only=True).round(1), inplace=True)
```

```python
demo.isna().sum()
```

```python
rmssd = pd.read_csv('original_RMSSD.csv')
rmssd.fillna(rmssd.mean(numeric_only=True).round(1), inplace=True)

rmssd.head()
```

```python
cols = rmssd.columns
nums = [1,3,4,5,8,9,10,13,14,15,18,19,20,23,24,25,28,29,30]

for i in nums:
    rmssd.drop(cols[i],axis=1,inplace=True)

rmssd = rmssd.add_prefix('rmssd_')
rmssd = rmssd.rename(columns={'rmssd_Subject_ID': 'ID'})
```

```python
rmssd.head()
```

```python
database = pd.merge(demo,rmssd, on="ID", how='inner')
```

```python
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
database.head()
```

#Making Calculation for best Exercise

```python
database.columns
```

```python
rmssd_anyb=[]
rmssd_om = []
rmssd_mm = []
rmssd_quit = []
rmssd_spine = []
rmssd_sh = []

# cols = ['ID', 'Gender', 'Age', 'Alchohol', 'Smoking', 'Non-Veg', 'Exercise',
#        'Medication', 'BMI', 'rmssd_PRE_ANYB', 'rmssd_POST_ANYB', 'rmssd_PRE_OM',
#        'rmssd_POST_OM', 'rmssd_PRE_SH', 'rmssd_POST_SH', 'rmssd_PRE_MM',
#        'rmssd_POST_MM', 'rmssd_PRE_QS', 'rmssd_POST_QS', 'rmssd_PRE_SR',
#        'rmssd_POST_SR']
       
for i in range(len(database['ID'])):

    val = database['rmssd_POST_ANYB'][i] - database['rmssd_PRE_ANYB'][i]
    rmssd_anyb.append(val)

for i in range(len(database['ID'])):

    val = database['rmssd_POST_OM'][i] - database['rmssd_PRE_OM'][i]
    rmssd_om.append(val)

for i in range(len(database['ID'])):

    val = database['rmssd_POST_SH'][i] - database['rmssd_PRE_SH'][i]
    rmssd_sh.append(val)

for i in range(len(database['ID'])):

    val = database['rmssd_POST_MM'][i] - database['rmssd_PRE_MM'][i]
    rmssd_mm.append(val)

for i in range(len(database['ID'])):

    val = database['rmssd_POST_QS'][i] - database['rmssd_PRE_QS'][i]
    rmssd_quit.append(val)

for i in range(len(database['ID'])):

    val = database['rmssd_POST_SR'][i] - database['rmssd_PRE_SR'][i]
    rmssd_spine.append(val)

database['rmssd_Anyb'] = rmssd_anyb
database['rmssd_om'] = rmssd_om
database['rmssd_mm'] = rmssd_mm
database['rmssd_sh'] = rmssd_sh
database['rmssd_quit'] = rmssd_quit
database['rmssd_spine'] =rmssd_spine
```

```python
selected_columns = ['rmssd_Anyb', 'rmssd_om', 'rmssd_mm','rmssd_sh', 'rmssd_quit', 'rmssd_spine']
col_dict = {'rmssd_Anyb':1, 'rmssd_om':2, 'rmssd_mm':3,'rmssd_sh':4, 'rmssd_quit':5, 'rmssd_spine':6}
# Function to find column number with maximum value in a row
def find_max_column(row):
    max_value = float('-inf')  # Initialize max value with negative infinity
    max_column = None          # Initialize max column as None

    for column in selected_columns:
        if row[column] > max_value:
            max_value = row[column]
            max_column = column

    return col_dict[max_column]

# Apply the function to each row and store the result in a new column
database['rmssd_best'] = database.apply(find_max_column, axis=1)
```

```python
drop_cols = ['rmssd_PRE_ANYB', 'rmssd_POST_ANYB',
       'rmssd_PRE_OM', 'rmssd_POST_OM', 'rmssd_PRE_SH', 'rmssd_POST_SH',
       'rmssd_PRE_MM', 'rmssd_POST_MM', 'rmssd_PRE_QS', 'rmssd_POST_QS',
       'rmssd_PRE_SR', 'rmssd_POST_SR', 'rmssd_Anyb', 'rmssd_om', 'rmssd_mm',
       'rmssd_sh', 'rmssd_quit', 'rmssd_spine']
database=database.drop(drop_cols,axis=1)
```
