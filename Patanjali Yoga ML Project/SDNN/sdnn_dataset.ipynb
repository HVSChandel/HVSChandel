```python
import pandas as pd
```

```python
demo = pd.read_csv('../Dataset - Demographic.csv')
demo.head()
```

```python
cols = demo.columns
nums = [1,9,10,11,13,14,15,16]

for i in nums:
    demo.drop(cols[i],axis=1,inplace=True)

demo = demo.rename(columns={'Subject_Id': 'ID'})
demo.fillna(demo.mean(numeric_only=True).round(1), inplace=True)
```

```python
demo.isna().sum()
```

```python
sdnn = pd.read_csv('original_SDNN.csv')
sdnn.fillna(sdnn.mean(numeric_only=True).round(1), inplace=True)

sdnn.head()
```

```python
cols = sdnn.columns
nums = [1,3,4,5,8,9,10,13,14,15,18,19,20,23,24,25,28,29,30]

for i in nums:
    sdnn.drop(cols[i],axis=1,inplace=True)

sdnn = sdnn.add_prefix('SDNN_')
sdnn = sdnn.rename(columns={'SDNN_Subject_ID': 'ID'})
```

```python
sdnn.head()
```

```python
database = pd.merge(demo,sdnn, on="ID", how='inner')
```

```python
database.columns
```

```python
sdnn_anyb=[]
sdnn_om = []
sdnn_mm = []
sdnn_quit = []
sdnn_spine = []
sdnn_sh = []

# cols = ['ID', 'Gender', 'Age', 'Alchohol', 'Smoking', 'Non-Veg', 'Exercise',
#        'Medication', 'BMI', 'SDNN_PRE_ANYB', 'SDNN_POST_ANYB', 'SDNN_PRE_OM',
#        'SDNN_POST_OM', 'SDNN_PRE_SH', 'SDNN_POST_SH', 'SDNN_PRE_MM',
#        'SDNN_POST_MM', 'SDNN_PRE_QS', 'SDNN_POST_QS', 'SDNN_PRE_SR',
#        'SDNN_POST_SR']
       
for i in range(len(database['ID'])):

    val = database['SDNN_POST_ANYB'][i] - database['SDNN_PRE_ANYB'][i]
    sdnn_anyb.append(val)

for i in range(len(database['ID'])):

    val = database['SDNN_POST_OM'][i] - database['SDNN_PRE_OM'][i]
    sdnn_om.append(val)

for i in range(len(database['ID'])):

    val = database['SDNN_POST_SH'][i] - database['SDNN_PRE_SH'][i]
    sdnn_sh.append(val)

for i in range(len(database['ID'])):

    val = database['SDNN_POST_MM'][i] - database['SDNN_PRE_MM'][i]
    sdnn_mm.append(val)

for i in range(len(database['ID'])):

    val = database['SDNN_POST_QS'][i] - database['SDNN_PRE_QS'][i]
    sdnn_quit.append(val)

for i in range(len(database['ID'])):

    val = database['SDNN_POST_SR'][i] - database['SDNN_PRE_SR'][i]
    sdnn_spine.append(val)

database['SDNN_Anyb'] = sdnn_anyb
database['SDNN_om'] = sdnn_om
database['SDNN_mm'] = sdnn_mm
database['SDNN_sh'] = sdnn_sh
database['SDNN_quit'] = sdnn_quit
database['SDNN_spine'] =sdnn_spine
```

```python
selected_columns = ['SDNN_Anyb', 'SDNN_om', 'SDNN_mm','SDNN_sh', 'SDNN_quit', 'SDNN_spine']
col_dict = {'SDNN_Anyb':1, 'SDNN_om':2, 'SDNN_mm':3,'SDNN_sh':4, 'SDNN_quit':5, 'SDNN_spine':6}
# Function to find column number with maximum value in a row
def find_max_column(row):
    max_value = float('-inf')  # Initialize max value with negative infinity
    max_column = None          # Initialize max column as None

    for column in selected_columns:
        if row[column] > max_value:
            max_value = row[column]
            max_column = column

    return col_dict[max_column]

# Apply the function to each row and store the result in a new column
database['SDNN_best'] = database.apply(find_max_column, axis=1)
```

```python
drop_cols = ['SDNN_PRE_ANYB', 'SDNN_POST_ANYB', 'SDNN_PRE_OM',
       'SDNN_POST_OM', 'SDNN_PRE_SH', 'SDNN_POST_SH', 'SDNN_PRE_MM',
       'SDNN_POST_MM', 'SDNN_PRE_QS', 'SDNN_POST_QS', 'SDNN_PRE_SR',
       'SDNN_POST_SR', 'SDNN_Anyb', 'SDNN_om', 'SDNN_mm', 'SDNN_sh',
       'SDNN_quit', 'SDNN_spine']
database=database.drop(drop_cols,axis=1)
```
