#Preprocessing


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/final_data_SDNN.csv')
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
df.columns
```

```python
X['Gender'] = X['Gender'].apply(lambda x: 0 if x == 'Male' else 1)
X['Alchohol'] = X['Alchohol'].apply(lambda x: 0 if x == 'No' else 1)
X['Smoking'] = X['Smoking'].apply(lambda x: 0 if x == 'No' else 1)
X['Non-Veg'] = X['Non-Veg'].apply(lambda x: 0 if x == 'No' else 1)
X['Exercise'] = X['Exercise'].apply(lambda x: 0 if x == 'No' else 1)
X['Medication'] = X['Medication'].apply(lambda x: 0 if x == 'No' else 1)
X.head()
```

```python
X.drop('ID',axis=1,inplace=True)
```

##OverSampler

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy="not majority")
X_res, y_res = ros.fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X = X_res.values
y = y_res.values
```

##Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test
```

```python
X_train.shape
```

##Feature-Scaling


```python
X_scaled_train = X_train
X_scaled_test = X_test

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

cols = [1,7]
scaler = StandardScaler()

scaler.fit(X_scaled_train[:, cols])

X_scaled_train[:, cols] = scaler.transform(X_scaled_train[:, cols])
X_scaled_test[:, cols] = scaler.transform(X_scaled_test[:, cols])
```

```python
X_train_normal = X_train
X_test_normal = X_test

from sklearn.preprocessing import Normalizer
nl_1 = Normalizer()

cols = [1,7]

norm = Normalizer()

norm.fit(X_train_normal[:, cols])

X_train_normal[:, cols] = norm.transform(X_train_normal[:, cols])
X_test_normal[:, cols] = norm.transform(X_test_normal[:, cols])
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

max_cm=0
best_ac=0
for i in range(100):
  n_esti = [3000,4000,5000,6000]
  ada_final = AdaBoostClassifier(estimator=GradientBoostingClassifier(n_estimators=5000), n_estimators=4000, learning_rate=0.01, random_state=0)
  ada_final.fit(X_train, y_train)
  y_pred = ada_final.predict(X_test)

  cm = confusion_matrix(y_test, y_pred)
  curr_ac = accuracy_score(y_test, y_pred)*100

  if(curr_ac>best_ac):
    max_cm = cm
    best_ac = curr_ac
print(max_cm)
print("\n")
print("Best Accuracy: {:.2f} %".format(best_ac))
```

##XGBoost

```python
import xgboost as xgb

from sklearn.model_selection import GridSearchCV
xgb_model = xgb.XGBClassifier()
y_train
parameters = [{'n_estimators': [10, 50, 100, 500, 1000, 1500, 2000],
               'learning_rate' : [0.1, 0.01, 0.001, 0.0001],
               'tree_method' : ['approx', 'hist']}]
grid_search = GridSearchCV(estimator = xgb_model,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search.fit(X_scaled_train, (y_train - 1))
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print("Best Accuracy: {:.2f} %".format(best_accuracy*100))
print("Best Parameters:", best_parameters)
```
