```python
import pandas as pd
```

```python
demo = pd.read_csv('../Dataset - Demographic.csv')
demo.head()
```

```python
cols = demo.columns
nums = [1,9,10,11,13,14,15,16]

for i in nums:
    demo.drop(cols[i],axis=1,inplace=True)

demo = demo.rename(columns={'Subject_Id': 'ID'})
demo.fillna(demo.mean(numeric_only=True).round(1), inplace=True)
```

```python
demo.isna().sum()
```

```python
HF = pd.read_csv('./original_HF(log).csv')
HF.fillna(HF.mean(numeric_only=True).round(1), inplace=True)

HF.head()
```

```python
cols = HF.columns
nums = [1,3,4,5,8,9,10,13,14,15,18,19,20,23,24,25,28,29,30]

for i in nums:
    HF.drop(cols[i],axis=1,inplace=True)

HF = HF.add_prefix('HF_')
HF = HF.rename(columns={'HF_Subject_ID': 'ID'})
```

```python
HF.head()
```

```python
database = pd.merge(demo,HF, on="ID", how='inner')
```

```python
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
database.head()
```

#Making Calculation for best Exercise

```python
database.columns
```

```python
HF_anyb=[]
HF_om = []
HF_mm = []
HF_quit = []
HF_spine = []
HF_sh = []


for i in range(len(database['ID'])):

    val = database['HF_ANYB_POST '][i] - database['HF_ANYB_PRE'][i]
    HF_anyb.append(val)

for i in range(len(database['ID'])):

    val = database['HF_OM_POST '][i] - database['HF_OM_PRE'][i]
    HF_om.append(val)

for i in range(len(database['ID'])):

    val = database['HF_SHA_POST '][i] - database['HF_SHA_PRE'][i]
    HF_sh.append(val)

for i in range(len(database['ID'])):

    val = database['HF_MM_POST '][i] - database['HF_MM_PRE'][i]
    HF_mm.append(val)

for i in range(len(database['ID'])):

    val = database['HF_QUIT_POST '][i] - database['HF_QUIT_PRE'][i]
    HF_quit.append(val)

for i in range(len(database['ID'])):

    val = database['HF_SPINE_POST '][i] - database['HF_SPINE_PRE'][i]
    HF_spine.append(val)

database['HF_Anyb'] = HF_anyb
database['HF_om'] = HF_om
database['HF_mm'] = HF_mm
database['HF_sh'] = HF_sh
database['HF_quit'] = HF_quit
database['HF_spine'] =HF_spine
```

```python
selected_columns = ['HF_Anyb', 'HF_om', 'HF_mm','HF_sh', 'HF_quit', 'HF_spine']
col_dict = {'HF_Anyb':1, 'HF_om':2, 'HF_mm':3,'HF_sh':4, 'HF_quit':5, 'HF_spine':6}
# Function to find column number with maximum value in a row
def find_max_column(row):
    max_value = float('-inf')  # Initialize max value with negative infinity
    max_column = None          # Initialize max column as None

    for column in selected_columns:
        if row[column] > max_value:
            max_value = row[column]
            max_column = column

    return col_dict[max_column]

# Apply the function to each row and store the result in a new column
database['HF_best'] = database.apply(find_max_column, axis=1)
```

```python
drop_cols = ['HF_ANYB_PRE', 'HF_ANYB_POST ', 'HF_OM_PRE',
       'HF_OM_POST ', 'HF_SHA_PRE', 'HF_SHA_POST ', 'HF_MM_PRE', 'HF_MM_POST ',
       'HF_QUIT_PRE', 'HF_QUIT_POST ', 'HF_SPINE_PRE', 'HF_SPINE_POST ',
       'HF_Anyb', 'HF_om', 'HF_mm', 'HF_sh', 'HF_quit', 'HF_spine']
database=database.drop(drop_cols,axis=1)
```
