```python
import pandas as pd
```

```python
demo = pd.read_csv('../Dataset - Demographic.csv')
demo.head()
```

```python
cols = demo.columns
nums = [1,9,10,11,13,14,15,16]

for i in nums:
    demo.drop(cols[i],axis=1,inplace=True)

demo = demo.rename(columns={'Subject_Id': 'ID'})
demo.fillna(demo.mean(numeric_only=True).round(1), inplace=True)
```

```python
demo.isna().sum()
```

```python
sti = pd.read_csv('./original_STI.csv')
sti.fillna(sti.mean(numeric_only=True).round(1), inplace=True)

sti.head()
```

```python
sti.isna().sum()
```

```python
cols = sti.columns
nums = [1,3,4,5,8,9,10,13,14,15,18,19,20,23,24,25,28,29,30]

for i in nums:
    sti.drop(cols[i],axis=1,inplace=True)

sti = sti.add_prefix('sti_')
sti = sti.rename(columns={'sti_Subject_ID': 'ID'})
```

```python
sti.head()
```

```python
database = pd.merge(demo,sti, on="ID", how='inner')
```

```python
sti_anyb=[]
sti_om = []
sti_mm = []
sti_QUIT = []
sti_spine = []
sti_sh = []

# cols = ['ID', 'Gender', 'Age', 'Alchohol', 'Smoking', 'Non-Veg', 'Exercise',
#        'Medication', 'BMI', 'sti_ANYB_PRE', 'sti_ANYB_POST  ', 'sti_OM_PRE',
#        'sti_OM_POST  ', 'sti_SHA_PRE', 'sti_SHA_POST  ', 'sti_MM_PRE',
#        'sti_MM_POST  ', 'sti_QUIT_PRE', 'sti_QUIT_POST  ', 'sti_SPINE_PRE',
#        'sti_SPINE_POST  ']
       
for i in range(len(database['ID'])):

    val = database['sti_ANYB_POST '][i] - database['sti_ANYB_PRE'][i]
    sti_anyb.append(val)

for i in range(len(database['ID'])):

    val = database['sti_OM_POST '][i] - database['sti_OM_PRE'][i]
    sti_om.append(val)

for i in range(len(database['ID'])):

    val = database['sti_SHA_POST '][i] - database['sti_SHA_PRE'][i]
    sti_sh.append(val)

for i in range(len(database['ID'])):

    val = database['sti_MM_POST '][i] - database['sti_MM_PRE'][i]
    sti_mm.append(val)

for i in range(len(database['ID'])):

    val = database['sti_QUIT_POST '][i] - database['sti_QUIT_PRE'][i]
    sti_QUIT.append(val)

for i in range(len(database['ID'])):

    val = database['sti_SPINE_POST '][i] - database['sti_SPINE_PRE'][i]
    sti_spine.append(val)

database['sti_Anyb'] = sti_anyb
database['sti_om'] = sti_om
database['sti_mm'] = sti_mm
database['sti_sh'] = sti_sh
database['sti_QUIT'] = sti_QUIT
database['sti_spine'] =sti_spine
```

```python
selected_columns = ['sti_Anyb', 'sti_om', 'sti_mm','sti_sh', 'sti_QUIT', 'sti_spine']
col_dict = {'sti_Anyb':1, 'sti_om':2, 'sti_mm':3,'sti_sh':4, 'sti_QUIT':5, 'sti_spine':6}
# Function to find column number with minimum value in a row
def find_min_column(row):
    min_value = float('inf')  # Initialize min value with negative infinity
    min_column = None          # Initialize min column as None

    for column in selected_columns:
        if row[column] < min_value:
            min_value = row[column]
            min_column = column

    return col_dict[min_column]

# Apply the function to each row and store the result in a new column
database['sti_best'] = database.apply(find_min_column, axis=1)
```
