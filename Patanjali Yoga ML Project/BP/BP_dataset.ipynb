```python
import pandas as pd
```

```python
demo = pd.read_csv('../Dataset - Demographic.csv')
demo.head()
```

```python
cols = demo.columns
nums = [1,9,10,11,13,14,15,16]

for i in nums:
    demo.drop(cols[i],axis=1,inplace=True)

demo = demo.rename(columns={'Subject_Id': 'ID'})
demo.fillna(demo.mean(numeric_only=True).round(1), inplace=True)
```

```python
bp = pd.read_csv('./original_BP_Systolic.csv')
bp.fillna(bp.mean(numeric_only=True).round(1), inplace=True)

bp.head()
```

```python
cols = bp.columns
nums = [1,2,3,4,6,7,8,10,11,12,14,15,16,18,19,20,22,23,24,26,27,28,30,31,32,34,35,36,38,39,40,42,43,44,46,47,48]

for i in nums:
    bp.drop(cols[i],axis=1,inplace=True)

bp = bp.add_prefix('BP_')
bp= bp.rename(columns={'BP_Subject_ID': 'ID'})
```

```python
database = pd.merge(demo,bp, on="ID", how='inner')
```

```python
selected_columns = ['BP_Anyb','BP_om', 'BP_mm', 'BP_sh', 'BP_quit', 'BP_spine']
col_dict = {'BP_Anyb':1,'BP_om':2, 'BP_mm':3, 'BP_sh':4, 'BP_quit':5, 'BP_spine':6}
# Function to find column number with maximum value in a row
def find_max_column(row):
    max_value = float('-inf')  # Initialize max value with negative infinity
    max_column = None          # Initialize max column as None

    for column in selected_columns:
        if row[column] > max_value:
            max_value = row[column]
            max_column = column

    return col_dict[max_column]

# Apply the function to each row and store the result in a new column
database['BP_best'] = database.apply(find_max_column, axis=1)
```
